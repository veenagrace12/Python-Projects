```python
import random

HANGMAN = [
"""
+---+
    |
    |
    |
    ===""", """
+---+
  | |
  O |
    |
    ===""", """
+---+
  | |
  O |
 /| |
    ===""", """
+---+
  | |
  O |
 /|\|
    ===""", """
+---+
  | |
  O |
 /|\|
  | ===""", """
 /   
+---+
  | |
  O |
 /|\|
  | ===""", """
 / \  
"""
]

words_list = ['HANGMAN', 'DAWNING', 'STATISTICS', 'PYTHON', 'ORACLE', 'DEVELOPER', 'ARROGANT', 'FORTUNATE']

word = random.choice(words_list).upper()

guessed_correctly = []
guessed_incorrectly = []
tries = 6
hangman_count = -1
while tries > 0:
    output = ''
    for letter in word:
        if letter in guessed_correctly:
            output += letter
        else:
            output += '_ '
    if output == word:
        break

    print('Guess the word: ',output)
    print(tries,' chances left')
    guess = input().upper()
    if guess in guessed_correctly or guess in guessed_incorrectly:
        print('Already guessed', guess)
    elif guess in word:
        print('Awesome! You guessed a correct letter!')
        guessed_correctly.append(guess)
    else:
        print('Sorry! You have guessed a wrong letter!')
        hangman_count = hangman_count + 1
        tries = tries - 1
        guessed_incorrectly.append(guess)
        print(HANGMAN[hangman_count])
if tries > 0:
    print('You guessed it right and you win!!!')
else:
    print('Sorry you guessed the wrong letter, try again.')
```
