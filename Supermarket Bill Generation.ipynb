```python

from datetime import datetime
name=input("Enter your name:")
lists='''
rice  ->  Rs 20/kg
sugar ->  Rs 30/kg
salt  ->  Rs 20/kg
oil   ->  Rs 80/litre
wheat -> Rs 110/kg
yipee ->  Rs 50/kg
boost ->  Rs 90/kg
onions-> Rs 85/kg
'''
#declaration
price=0
pricelist=[]
totalprice=0
Finalfinalprice=0
ilist=[]
qlist=[]
plist=[]

#rates for items
items={'rice':20,
'sugar':30,
'salt':20,
'oil':80,
'wheat':110,'yipee':50,'boost':90,'onions':85}
option=int(input("list of items, press 1:"))
if option==1:
    print(lists)
for i in range(len(items)):
    inp1=int(input("if you want to buy press 1 or 2 for exit:"))
    if inp1==2:
          break
    if inp1==1:
        item=input("Enter your items:")
        quantity=int(input("Enter quantity:"))
        if item in items.keys():
            price=quantity*(items[item])
            pricelist.append((item,quantity,items,price))
            totalprice+=price
            ilist.append(item)
            qlist.append(quantity)
            plist.append(price)
            gst=(totalprice*5)/100
            finalamount=gst+totalprice
        else:
            print("sorry the item is not available")
    else:
        print("you entered wrong number")
    inp=input("Bill the items, yes or no:")
    if inp=='yes':
        pass
        if finalamount!=0:
            print()
            print(25*"=","AV MART ",25*"=")
            print(28*" ","Vizag")
            print("Name:",name,30*" ","Date:",datetime.now())
            print(75*"-")
            print('S.no',"\t","\t",'Items',"\t","\t"'Quantity',"\t",'Price')
            for i in range(len(pricelist)):
                print(i,ilist[i],qlist[i],plist[i],sep='\t\t')
            print(75*"-")
            print(50*" ",'Totalamount:','Rs',totalprice)
            print('gstamount:',50*" ",'Rs',gst)
            print(75*"-")
            print(50*" ",'Finalamount:','Rs',finalamount)
            print(75*"-")
            print(50*" ","Thanks for visiting")
            print(75*"-")    
```
